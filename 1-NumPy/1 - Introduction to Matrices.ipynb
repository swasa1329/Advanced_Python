###### Addition & Multiplication of Matrices

```python
(1) Addition of matrices :

Matrices can be added only if they have the same dimensions.
```

```python
(2) Scalar multiplication :

 It multiplies every number in matrix with a scalar value.
```

```python
(3) Matrix Multiplication :

Matrices (other than the scalar) can be multiplied only if the number of columns 
of the left matrix is equals to the number of rows in the right matrix.

The multiplication of matrix is not commutative, it means A*B != B*A
```
