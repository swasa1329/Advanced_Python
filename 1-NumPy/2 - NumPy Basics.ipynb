###### What is NumPy ?

```python
NumPy is a scientific computing package (library) for python programming language.

Numpy is a powerful Python programming language library to solve numerical problems.
```

###### Meaning of Numpy:-

```python
Num stands for numerical and Py stands for Python programming language.

NumPy stands for numeric python which is a python package for the computation and processing 
of the multidimensional and single dimensional array elements.

NumPy supports large data in the form of a multidimensional array (vector and matrix).
```

```python
Representation of NumPy multidimensional arrays.
```

```python
The above figure shows one dimensional (1D), two dimensional (2D) and three dimensional (3D) NumPy array

One Dimensional NumPy array (1D): It means the collection of homogenous data in a single row (vector).

Two Dimensional NumPy arrays (2D): It means the collection of homogenous data in lists of a list (matrix).

Three Dimensional NumPy arrays (3D): It means the collection of homogenous data in lists of lists of a list (tensor).
```

###### Why NumPy array instead of Python List ?

```python
# Similarities between list and NumPy array:-
--------------------------------------------
1. Storing data - Both are used to store the data

2. Mutable

3. Can be Indexed

4. Slicing operation
```

```python
# Difference between list and NumPy array:-
-------------------------------------------
1. List :- we can store different data types in list. Ex:- [1, 20.5,"a",1j]
   NumPy Array:- We can only store similar data types in arrays. Ex:- [1,2,4,5]
        
2. Operations: we can perform all arthimetic operations on Numpy arrays, but will not be possible
    on lists. 
    
    Ex:- array([3,6,9,12])
        a/3
       we will get output as array([1.,2.,3.,4])
    
    List1 = [2,4,6,8]
    List1/2 - displays error. 
    
3. Numpy Array - to use it, we need to install NumPy package
   List :- it is a Built-in. no need to install seperately. 
```
