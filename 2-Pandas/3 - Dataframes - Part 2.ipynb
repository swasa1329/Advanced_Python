###### 2. Sum()

```python
The sum() method adds all values in each column and returns the sum for each column.

By specifying the column axis (axis='columns'), the sum() method searches column-wise and returns the sum of each row.

DataFrame.sum() function returns the sum of the values for the requested axis.
```

```python
# Create DataFrame

import pandas as pd

df1 = pd.DataFrame({'points': [18, 22, 19, 14, 14, 11, 20, 28],
                   'assists': [5, 7, 7, 9, 12, 9, 9, 4],
                   'rebounds': [11, 8, 10, 6, 6, 5, 9, 12]})

df2 = pd.DataFrame(df1)

df2
```

- Use DataFrame.sum() to get sum/total of a DataFrame for both rows and columns, 
<br>
- To get the total sum of columns use axis=1 param. 
<br>

- By default, this method takes axis=0 which means the summing of rows.

**1. Using DataFrame.sum() to Sum All Columns**

```python
# Example 1: Find Sum of All Columns

# The following code shows how to sum the values of the rows across all columns in the DataFrame:

df3 = df2.sum(axis=1)

print("Sum of all columns:-")

df3
```

```python
# Example 2: define new column that contains sum of all columns

df2['sum_stats'] = df2.sum(axis=1)

# view updated DataFrame

df2

# The sum_stats column contains the sum of the row values across all columns.

# For example, here’s how the values were calculated:

# Sum of row 0: 18 + 5 + 11 = 34
# Sum of row 1: 22 + 7 + 8 = 37
# Sum of row 2: 19 + 7 + 10 = 36
# And so on.
```

**2. Find Sum of Specific Columns**

```python
# Example 2: Find Sum of Specific Columns
    
# The following code shows how to sum the values of the rows across all columns in the DataFrame:

# specify the columns to sum

cols = ['points', 'assists']

#define new column that contains sum of specific columns

df2['sum_stats'] = df2[cols].sum(axis=1)

# view updated DataFrame

df2
```

**3. Using DataFrame.sum() to Sum of each row**

```python
df3 = df2.sum(axis=0)  # By default, it will take axis=0, if we didnt specify axis. 

print(df3)
```

  column labels, range index and memory consumption etc.
<br>
- The info() method in pandas return the summary of a dataframe. 
<br>

- info() should be the very first command, we run after uploading our data

==============================================================================<br>

**Syntax:-  DataFrame.info(verbose=None, buf=None, max_cols=None, memory_usage=None, show_counts=None, null_counts=None)**

**parameters:-**

**1. verbose:** It is of boolean type. Use it True when you want to get full information and false for less information.

**2. buf:** writable buffer, defaults to sys.stdout

**3. memory_usage:** Specifies whether total memory usage of the DataFrame elements (including the index) should be displayed.

**4. show_counts:** Whether to show the non-null counts.

**5. null_counts:** It is of boolean type

**Example 1:-Show the complete information about the dataframe.**

```python
# We can display the full information about the dataframe by passing the verbose =True
# as an argument of the info() function.

# We will get the following information.

# 1. class of the dataframe

# 2. number of rows and columns for the dataframe.

# 3. Type of the all columns

# 4. memory used by the dataframe.

# 5. Does column values contain null values or not?

import pandas as pd

import numpy as np

data = {"name":["Rob","Bam","Maya","Rahul"],"age":[23,25,26,32],"country":["USA","UK","France","Germany"]}

df = pd.DataFrame(data)

print(data)

print()

df.info(verbose=True)
```

#### 8. concat()

- Concat() method is used to concatenate/merge two or multiple pandas DataFrames across rows or columns. 
<br>
- When we concat() two pandas DataFrames on rows, it creates a new Dataframe containing all rows of two DataFrames.
<br>
- Basically it does append one DataFrame with another. 
<br>

- When we use concat() on columns, it performs the join operation.

**pandas concat() Key Points:-**

1. By default concat() method performs append operation meaning,<br>
   it appends each DataFrame at the end of the another DataFrame and creates a single DataFrame.
<br>
2. When we use concat() to join two DataFrames, it supports only inner and outer joins and by default it performs outer join.
<br>

3. Using concat,we can join or append multiple pandas DataFrames

**Syntax:-** pandas.concat(objs, axis=0, join='outer', ignore_index=False,<br>
             keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)


**1. objs:** a sequence of pandas objects to concatenate.<br>
**2. join:** optional parameter to define how to handle the indexes on the other axis. 
             The valid values are ‘inner’ and ‘outer’.<br>
**3.ignore_index:** if True, the indexes from the source objects will be ignored and
                     a sequence of indexes from 0,1,2…n will be assigned to the result.<br>
**4. keys:** a sequence to add an identifier to the result indexes. 
             It’s helpful in marking the source objects in the output.<br>
**5. levels:** a sequence to specify the unique levels to create multiindex.<br>
**6. names:** names for the levels in the resulting hierarchical index.<br>
**7. verify_integrity:** Check whether the new concatenated axis contains duplicates. It’s an expensive operation.<br>
**8. sort:** Sort non-concatenation axis if it is not already aligned when join is ‘outer’. Added in version 0.23.0.<br>
**9. copy:** if False, don’t copy data unnecessarily.

**Example 1:- To Concat Two DataFrames**

```python
import pandas as pd

df = pd.DataFrame({'Courses': ["Spark","PySpark","Python","pandas"],
                    'Fee' : [20000,25000,22000,24000]})

print("Dataframe1:- \n\n", df)

df1 = pd.DataFrame({'Courses': ["Pandas","Hadoop","Hyperion","Java"],
                    'Fee': [25000,25200,24500,24900]})

print("Dataframe1:- \n\n", df1)

# Using pandas.concat() to concat two DataFrames

data = [df, df1]

df2 = pd.concat(data)

df2

# Notice that the concatenation is performed row-wise i.e. 0-axis.
# Also, the indexes from the source DataFrame objects are preserved in the out
```

 **Example 2:- using the ignore_index=True param.**

```python
# Use pandas.concat() method to ignore_index

df2 = pd.concat([df, df1], ignore_index=True, sort=False)

df2
```

**Example 3:- To Join Two DataFrames**

```python
# pandas.concat() method is also used to join two pandas DataFrams on columns.

# In order to do so use axis=1, join='inner'. By default, pd.concat() is a row-wise outer join.

import pandas as pd

df = pd.DataFrame({'Courses':["Spark","PySpark","Python","pandas"],
                      'Fee' :[20000,25000,22000,24000]})

print("Dataframe1:-\n\n",df)

df1 = pd.DataFrame({'Duration':['30day','40days','35days','60days'],
                      'Discount':[1000,2300,2500,2000,]}) 

print("\n Dataframe2:-\n\n",df1)

# Using pandas.concat() to join concat two DataFrames

df2 = pd.concat([df, df1], axis=1, join='inner')

df2
```

**Example 4:- Concatenate Multiple DataFrames**

```python
import pandas as pd

df = pd.DataFrame({'Courses': ["Spark", "PySpark", "Python", "Pandas"],
                    'Fee' : ['20000', '25000', '22000', '24000']}) 
  
df1 = pd.DataFrame({'Courses': ["Unix", "Hadoop", "Hyperion", "Java"],
                    'Fee': ['25000', '25200', '24500', '24900']})
  
df2 = pd.DataFrame({'Duration':['30day','40days','35days','60days','55days'],
                    'Discount':[1000,2300,2500,2000,3000]})
  
# Appending multiple DataFrame

df3 = pd.concat([df, df1, df2],ignore_index=True, sort=False)

df3
```
