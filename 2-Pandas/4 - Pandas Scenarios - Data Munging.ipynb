###### 7. Merging

```python
import pandas as pd

zoo_eats = pd.DataFrame([['elephant','vegetables'],['tiger','meat'],
                         ['kangaroo','vegetables'],['zebra','vegetables'],['girafee','vegetables']], 
                        columns=['animal','food'])

zoo_eats
```

```python
df=zoo.merge(zoo_eats)

print("Merge of two data frames zoo & zoo_eats: -")

df
```

**1. Merge of two data frams with 'outer' parameter**

```python
df2=zoo.merge(zoo_eats, how='outer')

print("The Merge of two data frames with all Values:-")

df2
```

```python
If we observe in our dataframe zoo, we realize that there is no lion value in zoo_eats? 
Or that we don’t have any giraffes in zoo? 

When we did the merge above, by default, it was an INNER merge, 
so it filtered out giraffes and lions from the result table.

But there are cases in which, we do want to see these values in our joined dataframe.
```

**2. Merge of two data frams with Left Parameter**

```python
df3=zoo.merge(zoo_eats, how='left')

print("The Merge of two data frams with Left Parameter Values:-")

df3
```

```python
In this specific case, it would make more sense to keep lions in the table but not the giraffes.

With that, we could see all the animals in our zoo and we would have three food categories:-
vegetables, meat and NaN (which is basically “no information”). 

Keeping the giraffe line would be misleading and irrelevant, since we don’t have any
giraffes in our zoo anyway. 

That’s when merging with a how = 'left' parameter becomes handy!
```

**3. Merge of two data frams with Right Parameter**

```python
df4=zoo.merge(zoo_eats, how='right')

print("The Merge of two data frams with Right Parameter Values:-")

df4
```

###### 9. Resetting Index

```python
# When we have done a transformation on our dataframe, we have to re-index the rows.
# For that, we can use the reset_index() method.

rs=zoo.sort_values(by=['water_need'], ascending = False).reset_index()

print("Resetting index values:-")

rs
```

**Dropping old index**

```python
# As we can see, our new dataframe kept the old indexes, too.
# If we want to remove them, just add the drop = True parameter
```

```python
rs1=zoo.sort_values(by=['water_need'], ascending = False).reset_index(drop=True)

print("Resetting index values:-")

rs1
```

###### 10. Fillna()

```python
Pandas 'Fillna' - Replacing Nan with some meaning full name or adding a new name.

We have NaN values for lions. NaN itself can be really distracting, 
so usually we would like to replace it with something more meaningful.

In some cases, this can be a 0 value, or in other cases a specific string value,
but this time, we ’ll go with unknown. 

Let’s use the fillna() function, which basically finds and replaces all NaN values in our dataframe:
```

```python
f1=zoo.merge(zoo_eats, how='left').fillna('unknown')

print('Fill in NaN Values with Unknown:-')

f1
```

```python
f2=zoo.merge(zoo_eats, how='left').fillna('meat')

print('Fill in NaN Values with meet:-')

f2
```
