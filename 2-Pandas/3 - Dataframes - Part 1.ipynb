#### Dataframes

1. A Dataframe is a two-dimensional data structure in the form of table. 
   i.e., data is aligned in a tabular fashion in rows and columns.
   <br>
2. Two-dimentional data ==> (rows, columns) ==> table ==> like excel sheet.
   <br>
   
3. DataFrames can also handle data of all types.
   <br>
   


###### Syntax

**pd.DataFrame(data, index, columns)**

  **1. data**    ==> int, float, complext numbers, list, tuple, dictionary, etc. <br>
  **2. index**   ==> row labels <br>
  **3. columns** ==> column labels

###### Elements of a DataFrame

****There are 3 main elements of a dataframe****

****1) Data****

****2) Rows****

****3) Columns****

#### 8. Data Manipulation in Pandas (DataFrame)

 - Column Addition, Deletion, and Selection.
   <br>
 - Adding a Column in DataFrame.
   <br>
   
 - Adding multiple columns in a DataFrame.

```python
list_df = pd.DataFrame(mylist, index = [1,2,3,4,5], columns = ['Desi Food', 'Snakcs'])

print('Before addition')

list_df
```

**a) Adding a column in a dataframe is the same as adding another key/value pair in dictionary.**

```python
# Adding a new column 'Sea Food' to the dataframe.

list_df['Sea Food'] = ['Fish', 'Fish Kabab', 'Fish kurry', 'Fish pulao', 'Finger Fish']

print('After Addition')

list_df
```

```python
# Adding a new column 'price' to the dataframe

list_df['price'] = [1000,3500,4000,2000,1500]

list_df
```

**b) We can also add column using insert() function**

```python
# Adding a new column 'Deal' to dataframe with insert() function

list_df.insert(loc=4, column = 'Deal',value = [800,3000,3500,1500,1200])

list_df
```

**insert()** function

   **1. loc** ==> position, where you want to insert the new column. <br>

   **2. column** ==> name of the new column.<br>

   **3. value** ==> values that will be inserted in the new column

***c) Deleting a Column from a DataFrame using del keyword***

```python
# The easiest way to remove a column from a dataframe is to use del keyword

print('Before deleting')
list_df
```

```python
print('After deleting')

del list_df['price']
list_df
```

***d) Deleting a Column from a DataFrame using pop() function***

```python
# Adding a new column 'Deal' to dataframe with insert() function

list_df.insert(loc=4, column = 'Deal',value = [800,3000,3500,1500,1200])

print(list_df)

print("------------------")

# After deleting using pop() function

list_df.pop('Deal')

list_df
```

**e) Deleting a Column from a DataFrame using drop() function**

- The drop() method removes the specified row or column.
<br>

- By specifying the column axis (axis='columns'), the drop() method removes the specified column.
<br>

- By specifying the row axis (axis='index'), the drop() method removes the specified row.

```python
# Example 1:-

# Adding a new column 'Deal' to dataframe with insert() function

list_df = pd.DataFrame(mylist, index = [1,2,3,4,5], columns = ['Desi Food','Snacks'])

print(list_df)

print("------------------")

# After deleting using pop() function

d1=list_df.drop("Snacks",axis='columns')

d1
```

```python
# Example 2:-

import pandas as pd

data = {"name": ["Sally", "Mary", "John"],"age": [50, 40, 30],"qualified": [True, False, False]}

df = pd.DataFrame(data)

newdf = df.drop("age", axis='columns')

newdf
```

**- Selecting Multiple Columns**

**dataframe.columns can be used to select multple columns in a dataframe**

```python
import pandas as pd

# create a sample DataFrame

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

print(df)

# select columns A and B

df_AB = df[['A', 'B']]

df_AB
```

###### 9. Adding sum, difference, product, and division columns in our dataframe

```python
a = [1,3,5,7,9,11]
b = [2,4,6,8,10,12]
c = [2,3,5,7,11,13]

merger = list(zip(a,b,c))

print("List Values:-",merger)

mer_df = pd.DataFrame(merger, index = [1,2,3,4,5,6], columns = ['Odd', 'Even', 'Prime'])

mer_df
```

**1. Addition of the columns**

```python
mer_df['Sum'] = mer_df['Odd'] + mer_df['Even'] + mer_df['Prime']

mer_df
```

**2. Difference of the columns**

```python
mer_df['Difference'] = mer_df['Odd'] - mer_df['Even'] - mer_df['Prime']

mer_df
```

**3. Product of the columns**

```python
mer_df['Product'] = mer_df['Odd'] * mer_df['Even'] * mer_df['Prime']

mer_df
```

**4. Division of the columns**

```python
mer_df['Quotient'] = mer_df['Odd'] / mer_df['Even'] / mer_df['Prime']

mer_df
```
